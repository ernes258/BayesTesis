```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
datos=pd.read_csv('Casos_positivos_de_COVID-19_en_Colombia.csv')  
datos["Sexo"]=np.where(datos["Sexo"]=="f", "F", datos["Sexo"])
lista1=['ID de caso', 'Fecha de notificación',
       'Código DIVIPOLA departamento', 'Nombre departamento', 'Edad','Sexo', 'Tipo de contagio',
       'Ubicación del caso', 'Estado', 'Nombre del país', 'Recuperado','Fecha de inicio de síntomas', 'Fecha de muerte',
       'Fecha de diagnóstico', 'Fecha de recuperación','Pertenencia étnica']
datos=datos[lista1]
datos.head(5)
```

```python
datos["Nombre departamento"]=np.where(datos["Nombre departamento"]=="BARRANQUILLA", "ATLANTICO", datos["Nombre departamento"])
datos["Nombre departamento"]=np.where(datos["Nombre departamento"]=="STA MARTA D.E.", "MAGDALENA", datos["Nombre departamento"])
datos["Nombre departamento"]=np.where(datos["Nombre departamento"]=="CARTAGENA", "BOLIVAR", datos["Nombre departamento"])
datos["Código DIVIPOLA departamento"]=np.where(datos["Código DIVIPOLA departamento"]=="13,001", 13, datos["Código DIVIPOLA departamento"])
datos["Código DIVIPOLA departamento"]=np.where(datos["Código DIVIPOLA departamento"]=="47,001", 47, datos["Código DIVIPOLA departamento"])
datos["Código DIVIPOLA departamento"]=np.where(datos["Código DIVIPOLA departamento"]=="8,001", 8, datos["Código DIVIPOLA departamento"])
datos["Código DIVIPOLA departamento"] = datos["Código DIVIPOLA departamento"].astype(int)
#datos["Fecha de diagnóstico"]= pd.to_datetime(datos["Fecha de diagnóstico"])
#datos["Fecha de notificación"]= pd.to_datetime(datos["Fecha de notificación"])
#datos["Fecha de inicio de síntomas"]= pd.to_datetime(datos["Fecha de inicio de síntomas"])
#datos["Fecha de recuperación"]= pd.to_datetime(datos["Fecha de recuperación"])
#datos["Fecha de muerte"]= pd.to_datetime(datos["Fecha de muerte"])
datos.head(5)
```

```python
Agrupacion=pd.DataFrame(datos.groupby(["Fecha de notificación","Código DIVIPOLA departamento","Nombre departamento"]).size().reset_index(name='Casos'))
Agrupacion1=pd.DataFrame(datos.groupby(["Fecha de recuperación","Código DIVIPOLA departamento","Nombre departamento"]).size().reset_index(name='Casos'))
Agrupacion2=pd.DataFrame(datos.groupby(["Fecha de muerte","Código DIVIPOLA departamento","Nombre departamento"]).size().reset_index(name='Casos'))
Agrupacion2["Fecha de muerte"]= pd.to_datetime(Agrupacion2["Fecha de muerte"])
```

```python
Agrupacion.to_csv('Casos.csv')
Agrupacion1.to_csv('Recuperado.csv')
Agrupacion2.to_csv('Muerte.csv')
```

```python
# Creación DataFrame
list2=Agrupacion["Nombre departamento"].unique()
list2
df = pd.DataFrame(columns = ['Fecha', 'Departamento'])
for i in list2:
    consolidado=pd.DataFrame(pd.date_range(start=Agrupacion["Fecha de notificación"].min(), end=Agrupacion["Fecha de notificación"].max()))
    consolidado = consolidado.rename(columns={0:'Fecha'})
    consolidado["Fecha"]= pd.to_datetime(consolidado["Fecha"])
    consolidado["Departamento"]=i
    df = df.append(consolidado)
df=df.reset_index()

#Alimentar DataFrame
list1=[]
list3=[]
list4=[]
for i in range(len(df)):
    Date=df["Fecha"][i]
    Depto=df["Departamento"][i]
    Cases_ac=Agrupacion[(Agrupacion["Nombre departamento"]==Depto) & (Agrupacion["Fecha de notificación"]<=Date)]["Casos"].sum()
    Cases_re=Agrupacion1[(Agrupacion1["Nombre departamento"]==Depto) & (Agrupacion1["Fecha de recuperación"]<=Date)]["Casos"].sum()
    Cases_mu=Agrupacion2[(Agrupacion2["Nombre departamento"]==Depto) & (Agrupacion2["Fecha de muerte"]<=Date)]["Casos"].sum()
    
    list1.append(Cases_ac)
    list3.append(Cases_re)
    list4.append(Cases_mu)
    
list1=pd.DataFrame(list1, columns=["Casos"])
list3=pd.DataFrame(list3, columns=["Recuperados"])
list4=pd.DataFrame(list4, columns=["Muertos"])
df=pd.concat([df, list1], axis=1)
df=pd.concat([df, list3], axis=1)
df=pd.concat([df, list4], axis=1)
df["Casos"]=np.where((df["Recuperados"]+df["Muertos"])>df["Casos"], df["Recuperados"]+df["Muertos"], df["Casos"])
df
```

```python
pruebas=pd.read_csv('Pruebas_PCR_procesadas_de_COVID-19_en_Colombia__Departamental_.csv')
pruebas
```

```python
datos_prueba=pruebas.iloc[68:, :] # Primeras cinco columnas
datos_prueba
df = pd.DataFrame(columns = ['Fecha', 'Departamento'])
datos_prueba.head(5)
```

```python
df1 = pd.DataFrame(columns = ['Fecha', 'Pruebas', 'Departamento'])
lista_columnas=datos_prueba.columns
for i in range(2,39):
    dd=datos_prueba[["Fecha",lista_columnas[i]]]
    dd = dd.rename(columns={lista_columnas[i]:'Pruebas'})
    dd["Departamento"]=lista_columnas[i]
    df1=pd.concat([df1,dd]) 
df1=df1.reset_index()
df1
```

```python
df1["Departamento"]=df1["Departamento"].str.upper()
```

```python
df1["Departamento"]=np.where(df1["Departamento"]=="VALLE DEL CAUCA", "VALLE", df1["Departamento"])
df1["Departamento"]=np.where(df1["Departamento"]=="BARRANQUILLA", "ATLANTICO", df1["Departamento"])
df1["Departamento"]=np.where(df1["Departamento"]=="CARTAGENA", "BOLIVAR", df1["Departamento"])
df1["Departamento"]=np.where(df1["Departamento"]=="SANTA MARTA", "MAGDALENA", df1["Departamento"])
df1["Departamento"].unique()
```

```python
df2=pd.DataFrame(df1.groupby(["Fecha","Departamento"]).agg({'Pruebas': 'sum'}).reset_index())
df2["Fecha"]= pd.to_datetime(df2["Fecha"])
df2
```
